```python
# In[1]:
# ==============================================================================
# LIBRARIES AND SETUP
# ==============================================================================
# Install necessary libraries if you haven't already
# !pip install gymnasium torch stable-baselines3 pyyaml tqdm lightgbm

import numpy as np
import h5py
import yaml
import os
import torch
import gymnasium as gym
from gymnasium import spaces
from stable_baselines3 import TD3, PPO
from stable_baselines3.common.noise import NormalActionNoise
from stable_baselines3.common.callbacks import BaseCallback
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
import lightgbm as lgb
from scipy.stats import norm
import pickle
import warnings

# Suppress the specific warning about feature names
warnings.filterwarnings("ignore", category=UserWarning, message="X does not have valid feature names")

print("Libraries loaded successfully.")


# In[2]:
# ==============================================================================
# MASTER CONFIGURATION
# ==============================================================================
# --- Model Selection ---
# Define the window sizes for the PPO-Rolling models you want to test
ROLLING_WINDOW_SIZES = [6, 18, 36]

# Base models to run
MODELS_TO_RUN = [
    'LGBM', 
    'PPO', 
    # 'PPO-Augmented',
    'PPO-SNR-Augmented',
    # 'TD3',
    'MSC', 
    'CSM', 
    'TFG', 
    # 'TFL'
]
# Add the rolling models to the list automatically
for w in ROLLING_WINDOW_SIZES:
    MODELS_TO_RUN.append(f'PPO-Rolling-W{w}')
    MODELS_TO_RUN.append(f'PPO-Rolling-Augmented-W{w}')


# --- Data Loading & Training Parameters ---
TRAIN_TRIALS_PER_SNR = 5000
EVAL_TRIALS_PER_SNR = 200
RL_TOTAL_TIMESTEPS = 40000
LGBM_N_ESTIMATORS = 100
NEGATIVE_REWARD = -8.0

# --- General Configuration ---
config_path = '../config.yml'
with open(config_path, 'r') as f: config = yaml.safe_load(f)
p_conf = config['parameters']
hdf5_filepath = os.path.join('..', config['simulation']['hdf5_filepath'])
NUM_SIGNAL_FREQS = len(p_conf['signal_freqs'])
# We now evaluate on all SNRs since each contains signal and noise examples
ALL_SNRS = p_conf['snr_range'] 

print("Configuration loaded.")
print(f"Models to run: {MODELS_TO_RUN}")


# In[3]:
# ==============================================================================
# DATA LOADING AND PREPARATION (NEW LOGIC)
# ==============================================================================
def create_rolling_features(trial_data, window_size):
    """Helper function to create rolling sum features."""
    n_trials, n_steps, n_features = trial_data.shape
    rolling_features = np.zeros_like(trial_data)
    for i in range(n_steps):
        start_idx = max(0, i - window_size + 1)
        rolling_features[:, i, :] = np.sum(trial_data[:, start_idx:i+1, :], axis=1)
    return rolling_features

def load_and_prepare_data_v4(filepath, snr_levels, trials_per_snr, num_signal_freqs, window_sizes):
    """
    Loads data and prepares it for all model types with correct shuffling and augmentation.
    """
    X_sig_trials, X_noise_trials = [], []
    with h5py.File(filepath, 'r') as f:
        for snr in snr_levels:
            metrics_data = f[f'snr_{snr}']['metrics'][:trials_per_snr, :, :]
            X_sig_trials.append(metrics_data[:, :num_signal_freqs, :])
            X_noise_trials.append(metrics_data[:, num_signal_freqs:, :])

    X_sig = np.vstack(X_sig_trials)
    X_noise = np.vstack(X_noise_trials)
    
    # --- Create Standard (Flat) Dataset ---
    X_flat = np.vstack([X_sig.reshape(-1, 4), X_noise.reshape(-1, 4)])
    y_flat = np.concatenate([np.ones(X_sig.size // 4), np.zeros(X_noise.size // 4)])
    p_flat = np.random.permutation(len(X_flat))
    X_train_flat, y_train_flat = X_flat[p_flat], y_flat[p_flat]

    # --- Create Augmented Dataset (8 features) ---
    p_aug = np.random.permutation(len(X_sig))
    X_aug_trials = np.hstack([X_sig, X_noise[p_aug]])
    X_train_aug = X_aug_trials.reshape(-1, 8)
    y_train_aug = np.ones(len(X_train_aug))

    # --- Create SNR-Augmented Dataset (5 features) ---
    X_train_snr = np.hstack([X_train_flat, X_train_flat[:, 3:4]])
    y_train_snr = y_train_flat

    # --- Create Rolling and Rolling-Augmented Datasets ---
    rolling_datasets = {}
    rolling_aug_datasets = {}
    for w in window_sizes:
        X_rolling_sig = create_rolling_features(X_sig, w)
        X_rolling_noise = create_rolling_features(X_noise, w)
        
        # Standard Rolling
        X_rolling = np.vstack([X_rolling_sig.reshape(-1, 4), X_rolling_noise.reshape(-1, 4)])
        y_rolling = np.concatenate([np.ones(X_rolling_sig.size // 4), np.zeros(X_rolling_noise.size // 4)])
        p_roll = np.random.permutation(len(X_rolling))
        rolling_datasets[w] = (X_rolling[p_roll], y_rolling[p_roll])
        
        # Rolling Augmented
        p_roll_aug = np.random.permutation(len(X_rolling_sig))
        X_roll_aug_trials = np.hstack([X_rolling_sig, X_rolling_noise[p_roll_aug]])
        X_train_roll_aug = X_roll_aug_trials.reshape(-1, 8)
        y_train_roll_aug = np.ones(len(X_train_roll_aug))
        rolling_aug_datasets[w] = (X_train_roll_aug, y_train_roll_aug)

    return (X_train_flat, y_train_flat), (X_train_aug, y_train_aug), (X_train_snr, y_train_snr), rolling_datasets, rolling_aug_datasets

# Load distinct datasets for training
print("--- Loading Data for Training ---")
(X_train_flat, y_train_flat), (X_train_aug, y_train_aug), (X_train_snr, y_train_snr), rolling_train_data, rolling_aug_train_data = load_and_prepare_data_v4(
    hdf5_filepath, ALL_SNRS, TRAIN_TRIALS_PER_SNR, NUM_SIGNAL_FREQS, ROLLING_WINDOW_SIZES
)
print(f"Standard training data shape: X={X_train_flat.shape}")
print(f"Augmented training data shape: X={X_train_aug.shape}")
print(f"SNR-Augmented training data shape: X={X_train_snr.shape}")
for w, (x, y) in rolling_train_data.items(): print(f"Rolling (W={w}) training data shape: X={x.shape}")
for w, (x, y) in rolling_aug_train_data.items(): print(f"Rolling-Augmented (W={w}) training data shape: X={x.shape}")


# In[4]:
# ==============================================================================
# GYMNASIUM ENVIRONMENTS FOR RL
# ==============================================================================
class SignalDetectionEnv(gym.Env):
    """Standard environment for non-recurrent agents."""
    def __init__(self, features, labels, negative_reward=-1.0):
        super().__init__()
        self.features, self.labels = features, labels
        self.negative_reward = negative_reward
        self.current_step = 0
        self.action_space = spaces.Box(low=-1, high=1, shape=(1,), dtype=np.float32)
        self.observation_space = spaces.Box(low=-np.inf, high=np.inf, shape=(features.shape[1],), dtype=np.float32)

    def step(self, action):
        predicted_label = 1 if action[0] > 0 else 0
        reward = 1.0 if predicted_label == self.labels[self.current_step] else self.negative_reward
        self.current_step += 1
        terminated = self.current_step >= len(self.features)
        next_obs = self.features[self.current_step] if not terminated else np.zeros(self.observation_space.shape)
        return next_obs, reward, terminated, False, {}

    def reset(self, seed=None, options=None):
        super().reset(seed=seed)
        self.current_step = 0
        return self.features[0], {}

# Create the training environments
train_env = SignalDetectionEnv(X_train_flat, y_train_flat, negative_reward=NEGATIVE_REWARD)
train_env_aug = SignalDetectionEnv(X_train_aug, y_train_aug, negative_reward=NEGATIVE_REWARD)
train_env_snr = SignalDetectionEnv(X_train_snr, y_train_snr, negative_reward=NEGATIVE_REWARD)
train_envs_rolling = {w: SignalDetectionEnv(x, y, negative_reward=NEGATIVE_REWARD) for w, (x, y) in rolling_train_data.items()}
train_envs_rolling_aug = {w: SignalDetectionEnv(x, y, negative_reward=NEGATIVE_REWARD) for w, (x, y) in rolling_aug_train_data.items()}
print("Gymnasium environments created.")


# In[5]:
# ==============================================================================
# MODEL TRAINING
# ==============================================================================
class ORDTester:
    def __init__(self, desired_alpha=0.05):
        self.desired_alpha = desired_alpha
        self.feature_names = ['MSC', 'CSM', 'TFG', 'TFL']
        self.thresholds = {}
    def train(self, noise_features):
        for i, key in enumerate(self.feature_names):
            self.thresholds[key] = np.percentile(noise_features[:, i], (1 - self.desired_alpha) * 100)
    def predict(self, features):
        decisions = {}
        for i, name in enumerate(self.feature_names):
            decisions[name] = (features[:, i] > self.thresholds[name]).astype(int)
        return decisions

trained_models = {}
policy_kwargs = dict(net_arch=dict(pi=[64, 64], vf=[64, 64]))

if 'LGBM' in MODELS_TO_RUN:
    print("\n--- Training LightGBM ---")
    lgbm_model = lgb.LGBMClassifier(n_estimators=LGBM_N_ESTIMATORS, random_state=42)
    lgbm_model.fit(X_train_flat, y_train_flat)
    trained_models['LGBM'] = lgbm_model

if 'PPO' in MODELS_TO_RUN:
    print("\n--- Training PPO (4 Features) ---")
    ppo_model = PPO("MlpPolicy", train_env, policy_kwargs=policy_kwargs, n_steps=2048, verbose=0)
    ppo_model.learn(total_timesteps=RL_TOTAL_TIMESTEPS, progress_bar=True)
    trained_models['PPO'] = ppo_model
    
if 'PPO-Augmented' in MODELS_TO_RUN:
    print("\n--- Training PPO-Augmented (8 Features) ---")
    ppo_aug_model = PPO("MlpPolicy", train_env_aug, policy_kwargs=policy_kwargs, n_steps=2048, verbose=0)
    ppo_aug_model.learn(total_timesteps=RL_TOTAL_TIMESTEPS, progress_bar=True)
    trained_models['PPO-Augmented'] = ppo_aug_model

if 'PPO-SNR-Augmented' in MODELS_TO_RUN:
    print("\n--- Training PPO-SNR-Augmented (5 Features) ---")
    ppo_snr_model = PPO("MlpPolicy", train_env_snr, policy_kwargs=policy_kwargs, n_steps=2048, verbose=1)
    ppo_snr_model.learn(total_timesteps=RL_TOTAL_TIMESTEPS, progress_bar=True)
    trained_models['PPO-SNR-Augmented'] = ppo_snr_model

for w in ROLLING_WINDOW_SIZES:
    model_name = f'PPO-Rolling-W{w}'
    if model_name in MODELS_TO_RUN:
        print(f"\n--- Training {model_name} ---")
        model = PPO("MlpPolicy", train_envs_rolling[w], policy_kwargs=policy_kwargs, n_steps=2048, verbose=1)
        model.learn(total_timesteps=RL_TOTAL_TIMESTEPS, progress_bar=True)
        trained_models[model_name] = model

    model_name_aug = f'PPO-Rolling-Augmented-W{w}'
    if model_name_aug in MODELS_TO_RUN:
        print(f"\n--- Training {model_name_aug} ---")
        model = PPO("MlpPolicy", train_envs_rolling_aug[w], policy_kwargs=policy_kwargs, n_steps=2048, verbose=1)
        model.learn(total_timesteps=RL_TOTAL_TIMESTEPS, progress_bar=True)
        trained_models[model_name_aug] = model

ord_tester = ORDTester()
ord_tester.train(X_train_flat[y_train_flat == 0])
trained_models['ORD'] = ord_tester
print("All models trained.")

# In[6]:
# ==============================================================================
# FULL EVALUATION (NEW LOGIC)
# ==============================================================================
print("\n--- Starting Full Evaluation ---")

def predict_with_model(model_name, model, features):
    """A wrapper to handle different predict method signatures."""
    if "PPO" in model_name or "TD3" in model_name:
        preds, _ = model.predict(features, deterministic=True)
        preds[np.isnan(preds)] = 0
        return preds > 0
    elif "LGBM" in model_name:
        return model.predict(features)
    return None # Should not happen

results = {model: {'pd': [], 'fp': []} for model in MODELS_TO_RUN}

with h5py.File(hdf5_filepath, 'r') as f:
    for snr in tqdm(ALL_SNRS, desc="Evaluating across SNRs"):
        metrics_data = f[f'snr_{snr}']['metrics'][:EVAL_TRIALS_PER_SNR, :, :]
        
        signal_features_trials = metrics_data[:, :NUM_SIGNAL_FREQS, :]
        noise_features_trials = metrics_data[:, NUM_SIGNAL_FREQS:, :]
        
        signal_features_flat = signal_features_trials.reshape(-1, 4)
        noise_features_flat = noise_features_trials.reshape(-1, 4)

        # --- Evaluate PD on signal features ---
        for model_name, model in trained_models.items():
            if model_name == 'ORD' or model_name not in MODELS_TO_RUN: continue
            
            if model_name == 'PPO-Augmented':
                p_eval = np.random.permutation(len(noise_features_trials))
                eval_features = np.hstack([signal_features_trials, noise_features_trials[p_eval]]).reshape(-1, 8)
            elif model_name == 'PPO-SNR-Augmented':
                eval_features = np.hstack([signal_features_flat, signal_features_flat[:, 3:4]])
            elif 'PPO-Rolling-Augmented' in model_name:
                w = int(model_name.split('-W')[-1])
                rolling_sig = create_rolling_features(signal_features_trials, w)
                rolling_noise = create_rolling_features(noise_features_trials, w)
                p_eval = np.random.permutation(len(rolling_noise))
                eval_features = np.hstack([rolling_sig, rolling_noise[p_eval]]).reshape(-1, 8)
            elif 'PPO-Rolling' in model_name:
                w = int(model_name.split('-W')[-1])
                eval_features = create_rolling_features(signal_features_trials, w).reshape(-1, 4)
            else: # Standard models
                eval_features = signal_features_flat
            
            preds = predict_with_model(model_name, model, eval_features)
            results[model_name]['pd'].append(np.mean(preds))

        # Evaluate ORD models on signal features
        ord_preds = trained_models['ORD'].predict(signal_features_flat)
        for det in ['MSC', 'CSM', 'TFG', 'TFL']:
            if det in MODELS_TO_RUN: results[det]['pd'].append(np.mean(ord_preds[det]))

        # --- Evaluate FP on noise features ---
        for model_name, model in trained_models.items():
            if model_name == 'ORD' or model_name not in MODELS_TO_RUN: continue
            
            if model_name == 'PPO-Augmented':
                p_eval = np.random.permutation(len(signal_features_trials))
                eval_features = np.hstack([noise_features_trials, signal_features_trials[p_eval]]).reshape(-1, 8)
            elif model_name == 'PPO-SNR-Augmented':
                eval_features = np.hstack([noise_features_flat, noise_features_flat[:, 3:4]])
            elif 'PPO-Rolling-Augmented' in model_name:
                w = int(model_name.split('-W')[-1])
                rolling_sig = create_rolling_features(signal_features_trials, w)
                rolling_noise = create_rolling_features(noise_features_trials, w)
                p_eval = np.random.permutation(len(rolling_sig))
                eval_features = np.hstack([rolling_noise, rolling_sig[p_eval]]).reshape(-1, 8)
            elif 'PPO-Rolling' in model_name:
                w = int(model_name.split('-W')[-1])
                eval_features = create_rolling_features(noise_features_trials, w).reshape(-1, 4)
            else: # Standard models
                eval_features = noise_features_flat
            
            preds = predict_with_model(model_name, model, eval_features)
            results[model_name]['fp'].append(np.mean(preds))
            
        ord_preds = trained_models['ORD'].predict(noise_features_flat)
        for det in ['MSC', 'CSM', 'TFG', 'TFL']:
            if det in MODELS_TO_RUN: results[det]['fp'].append(np.mean(ord_preds[det]))

print("Evaluation complete.")


# In[7]:
# ==============================================================================
# PLOT 2: MEAN PERFORMANCE COMPARISON
# ==============================================================================
print("\n--- Generating Mean Performance Comparison Plot ---")
plt.figure(figsize=(20, 9))

model_styles = {
    'RL': {'cmap': plt.get_cmap('Blues'), 'markers': ['o', 's', 'v', '^', '<', '>', 'p']},
    'Supervised': {'cmap': plt.get_cmap('Greens'), 'markers': ['D']},
    'Statistical': {'cmap': plt.get_cmap('Purples'), 'markers': ['X', 'P', '*', 'H']}
}
rl_models = [m for m in MODELS_TO_RUN if 'PPO' in m or 'TD3' in m]
sup_models = [m for m in MODELS_TO_RUN if 'LGBM' in m]
stat_models = [m for m in MODELS_TO_RUN if m in ['MSC', 'CSM', 'TFG', 'TFL']]

def get_style(model_name):
    if model_name in rl_models:
        idx = rl_models.index(model_name)
        return model_styles['RL']['cmap'](0.4 + 0.6 * idx / len(rl_models)), model_styles['RL']['markers'][idx % len(model_styles['RL']['markers'])]
    if model_name in sup_models:
        idx = sup_models.index(model_name)
        return model_styles['Supervised']['cmap'](0.6), model_styles['Supervised']['markers'][idx % len(model_styles['Supervised']['markers'])]
    if model_name in stat_models:
        idx = stat_models.index(model_name)
        return model_styles['Statistical']['cmap'](0.4 + 0.6 * idx / len(stat_models)), model_styles['Statistical']['markers'][idx % len(model_styles['Statistical']['markers'])]
    return 'k', '.'

for model_name in MODELS_TO_RUN:
    color, marker = get_style(model_name)
    
    # Plot PD vs SNR
    plt.plot(ALL_SNRS, results[model_name]['pd'], marker=marker, linestyle='-', color=color, label=f'{model_name} PD')
    
    # Plot FPR vs SNR
    plt.plot(ALL_SNRS, results[model_name]['fp'], marker=marker, linestyle='--', color=color, label=f'{model_name} FPR')

plt.title('Detector Performance Comparison vs. SNR', fontsize=16)
plt.xlabel('SNR (dB)', fontsize=12)
plt.ylabel('Probability', fontsize=12)
plt.grid(True, which='both', linestyle=':')
plt.ylim([-0.05, 1.05])
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()
plt.show()

# In[8]:
# ==============================================================================
# PLOT 2: FPR COMPARISON
# ==============================================================================
print("\n--- Generating FPR vs. SNR Comparison Plot ---")
plt.figure(figsize=(14, 9))

for model_name in MODELS_TO_RUN:
    color, marker = get_style(model_name)
    plt.plot(ALL_SNRS, results[model_name]['fp'], marker=marker, linestyle='--', color=color, label=f'{model_name} FPR')

plt.title('Detector FPR Comparison vs. SNR', fontsize=16)
plt.xlabel('SNR (dB)', fontsize=12)
plt.ylabel('False Positive Rate (FPR)', fontsize=12)
plt.grid(True, which='both', linestyle=':')
# plt.ylim([-0.05, max(0.2, np.max([v['fp'] for v in results.values()]) * 1.2) if results else 0.2])
plt.ylim([0, 0.15])
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()
plt.show()
```
