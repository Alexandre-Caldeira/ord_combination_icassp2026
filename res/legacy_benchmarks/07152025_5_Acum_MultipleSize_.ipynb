```python
# In[1]:
# ==============================================================================
# LIBRARIES AND SETUP
# ==============================================================================
# Install necessary libraries if you haven't already
# !pip install gymnasium torch stable-baselines3 pyyaml tqdm lightgbm

import numpy as np
import h5py
import yaml
import os
import torch
import gymnasium as gym
from gymnasium import spaces
from stable_baselines3 import TD3, PPO
from stable_baselines3.common.noise import NormalActionNoise
from stable_baselines3.common.callbacks import BaseCallback
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
import lightgbm as lgb
from scipy.stats import norm
import pickle

print("Libraries loaded successfully.")


# In[2]:
# ==============================================================================
# MASTER CONFIGURATION
# ==============================================================================
# --- Model Selection ---
# Define the window sizes for the PPO-Rolling models you want to test
ROLLING_WINDOW_SIZES = [6, 8, 16]

# Base models to run
MODELS_TO_RUN = [
    'LGBM', 
    'PPO', 
    'PPO-Augmented',
    # 'TD3',
    'MSC', 
    # 'CSM', 
    'TFG', 
    # 'TFL'
]
# Add the rolling models to the list automatically
for w in ROLLING_WINDOW_SIZES:
    MODELS_TO_RUN.append(f'PPO-Rolling-W{w}')


# --- Data Loading Parameters ---
# How many trials to load from each SNR group for the TRAINING dataset
TRAIN_TRIALS_PER_SNR = 2000
# How many trials to load from each SNR group for the FINAL EVALUATION dataset
EVAL_TRIALS_PER_SNR = 1000

# --- Model Training Parameters ---
# Total steps to train the RL agents for
RL_TOTAL_TIMESTEPS = 20000
# Number of estimators for the LightGBM model
LGBM_N_ESTIMATORS = 100
# Reward for an incorrect decision in the RL environment
NEGATIVE_REWARD = -5.0

# --- General Configuration ---
# Load project config to get file paths and parameters
config_path = '../config.yml'
with open(config_path, 'r') as f:
    config = yaml.safe_load(f)

s_conf = config['simulation']
p_conf = config['parameters']
hdf5_filepath = os.path.join('..', s_conf['hdf5_filepath'])
NOISE_SNR = min(p_conf['snr_range'])
SIGNAL_SNRS = [snr for snr in p_conf['snr_range'] if snr != NOISE_SNR]
NUM_FREQS = len(p_conf['signal_freqs'])

print("Configuration loaded.")
print(f"Models to run: {MODELS_TO_RUN}")


# In[3]:
# ==============================================================================
# DATA LOADING AND PREPARATION
# ==============================================================================

def create_rolling_features(trial_data, window_size):
    """Helper function to create rolling sum features."""
    n_trials, n_steps, n_features = trial_data.shape
    rolling_features = np.zeros_like(trial_data)
    for i in range(n_steps):
        start_idx = max(0, i - window_size + 1)
        # Sum the features in the window along the time-step axis
        rolling_features[:, i, :] = np.sum(trial_data[:, start_idx:i+1, :], axis=1)
    return rolling_features

def load_and_prepare_data(filepath, noise_snr, signal_snrs, trials_per_snr, num_freqs, window_sizes):
    """
    Loads data and prepares it for all model types.
    """
    X_all_trials, y_all_trials = [], []
    with h5py.File(filepath, 'r') as f:
        # Load noise samples
        noise_trials = f[f'snr_{noise_snr}']['metrics'][:trials_per_snr, :, :]
        X_all_trials.append(noise_trials)
        y_all_trials.append(np.zeros((len(noise_trials), num_freqs)))
        
        # Load signal samples, balanced across SNRs
        total_signal_trials_needed = len(noise_trials)
        trials_per_snr_group = max(1, total_signal_trials_needed // len(signal_snrs))
        
        signal_trials_list = []
        for snr in signal_snrs:
            data = f[f'snr_{snr}']['metrics'][:trials_per_snr_group, :, :]
            signal_trials_list.append(data)
            
        X_signal_trials = np.vstack(signal_trials_list)
        X_signal_trials = X_signal_trials[:total_signal_trials_needed]
        
        X_all_trials.append(X_signal_trials)
        y_all_trials.append(np.ones((len(X_signal_trials), num_freqs)))

    # Combine and shuffle data at the trial level
    X_trials = np.vstack(X_all_trials)
    y_trials = np.vstack(y_all_trials)
    p = np.random.permutation(len(X_trials))
    X_trials, y_trials = X_trials[p], y_trials[p]

    # --- Create Standard (Flat) Dataset ---
    X_flat = X_trials.reshape(-1, 4)
    y_flat = y_trials.flatten()

    # --- Create Augmented Dataset (8 features) ---
    X_augmented_list = []
    for i in range(num_freqs):
        noise_ref_idx = (i + num_freqs // 2) % num_freqs
        augmented_features = np.hstack([X_trials[:, i, :], X_trials[:, noise_ref_idx, :]])
        X_augmented_list.append(augmented_features)
    X_aug = np.stack(X_augmented_list, axis=1).reshape(-1, 8)
    y_aug = y_trials.flatten()

    # --- Create Rolling Window Datasets ---
    rolling_datasets = {}
    for w in window_sizes:
        X_rolling_trials = create_rolling_features(X_trials, w)
        X_rolling = X_rolling_trials.reshape(-1, 4)
        y_rolling = y_trials.flatten()
        rolling_datasets[w] = (X_rolling, y_rolling)

    return (X_flat, y_flat), (X_aug, y_aug), rolling_datasets

# Load distinct datasets for training
print("--- Loading Data for Training ---")
(X_train_flat, y_train_flat), (X_train_aug, y_train_aug), rolling_train_data = load_and_prepare_data(
    hdf5_filepath, NOISE_SNR, SIGNAL_SNRS, TRAIN_TRIALS_PER_SNR, NUM_FREQS, ROLLING_WINDOW_SIZES
)
print(f"Standard training data shape: X={X_train_flat.shape}")
print(f"Augmented training data shape: X={X_train_aug.shape}")
for w, (x, y) in rolling_train_data.items():
    print(f"Rolling (W={w}) training data shape: X={x.shape}")


# In[4]:
# ==============================================================================
# GYMNASIUM ENVIRONMENTS FOR RL
# ==============================================================================

class SignalDetectionEnv(gym.Env):
    """Standard environment for non-recurrent agents."""
    def __init__(self, features, labels, negative_reward=-1.0):
        super().__init__()
        self.features, self.labels = features, labels
        self.negative_reward = negative_reward
        self.current_step = 0
        self.action_space = spaces.Box(low=-1, high=1, shape=(1,), dtype=np.float32)
        self.observation_space = spaces.Box(low=-np.inf, high=np.inf, shape=(features.shape[1],), dtype=np.float32)

    def step(self, action):
        predicted_label = 1 if action[0] > 0 else 0
        reward = 1.0 if predicted_label == self.labels[self.current_step] else self.negative_reward
        self.current_step += 1
        terminated = self.current_step >= len(self.features)
        next_obs = self.features[self.current_step] if not terminated else np.zeros(self.observation_space.shape)
        return next_obs, reward, terminated, False, {}

    def reset(self, seed=None, options=None):
        super().reset(seed=seed)
        self.current_step = 0
        return self.features[0], {}

# Create the training environments
train_env = SignalDetectionEnv(X_train_flat, y_train_flat, negative_reward=NEGATIVE_REWARD)
train_env_aug = SignalDetectionEnv(X_train_aug, y_train_aug, negative_reward=NEGATIVE_REWARD)
train_envs_rolling = {w: SignalDetectionEnv(x, y, negative_reward=NEGATIVE_REWARD) for w, (x, y) in rolling_train_data.items()}
print("Gymnasium environments created.")


# In[5]:
# ==============================================================================
# MODEL TRAINING
# ==============================================================================

class ORDTester:
    """A simplified tester for threshold-based detectors."""
    def __init__(self, desired_alpha=0.05):
        self.desired_alpha = desired_alpha
        self.feature_names = ['MSC', 'CSM', 'TFG', 'TFL']
        self.thresholds = {}

    def train(self, noise_features):
        print("Training ORDTester by calculating thresholds...")
        for i, key in enumerate(self.feature_names):
            self.thresholds[key] = np.percentile(noise_features[:, i], (1 - self.desired_alpha) * 100)
        print("Thresholds calculated.")

    def predict(self, features):
        decisions = {}
        for i, name in enumerate(self.feature_names):
            decisions[name] = (features[:, i] > self.thresholds[name]).astype(int)
        return decisions

# --- Train all selected models ---
trained_models = {}

if 'LGBM' in MODELS_TO_RUN:
    print("\n--- Training LightGBM ---")
    lgbm_model = lgb.LGBMClassifier(n_estimators=LGBM_N_ESTIMATORS, random_state=42)
    lgbm_model.fit(X_train_flat, y_train_flat)
    trained_models['LGBM'] = lgbm_model
    print("LGBM training complete.")

if 'PPO' in MODELS_TO_RUN:
    print("\n--- Training PPO (4 Features) ---")
    ppo_model = PPO("MlpPolicy", train_env, verbose=0, tensorboard_log="./ppo_tensorboard/")
    ppo_model.learn(total_timesteps=RL_TOTAL_TIMESTEPS, progress_bar=True)
    trained_models['PPO'] = ppo_model
    print("PPO training complete.")
    
if 'PPO-Augmented' in MODELS_TO_RUN:
    print("\n--- Training PPO-Augmented (8 Features) ---")
    ppo_aug_model = PPO("MlpPolicy", train_env_aug, verbose=0, tensorboard_log="./ppo_aug_tensorboard/")
    ppo_aug_model.learn(total_timesteps=RL_TOTAL_TIMESTEPS, progress_bar=True)
    trained_models['PPO-Augmented'] = ppo_aug_model
    print("PPO-Augmented training complete.")

for w in ROLLING_WINDOW_SIZES:
    model_name = f'PPO-Rolling-W{w}'
    if model_name in MODELS_TO_RUN:
        print(f"\n--- Training {model_name} ---")
        ppo_rolling_model = PPO("MlpPolicy", train_envs_rolling[w], verbose=1, tensorboard_log=f"./{model_name}_tensorboard/")
        ppo_rolling_model.learn(total_timesteps=RL_TOTAL_TIMESTEPS, progress_bar=True)
        trained_models[model_name] = ppo_rolling_model
        print(f"{model_name} training complete.")

if 'TD3' in MODELS_TO_RUN:
    print("\n--- Training TD3 ---")
    n_actions = train_env.action_space.shape[-1]
    action_noise = NormalActionNoise(mean=np.zeros(n_actions), sigma=0.1 * np.ones(n_actions))
    td3_model = TD3("MlpPolicy", train_env, action_noise=action_noise, verbose=0, tensorboard_log="./td3_tensorboard/")
    td3_model.learn(total_timesteps=RL_TOTAL_TIMESTEPS, progress_bar=True)
    trained_models['TD3'] = td3_model
    print("TD3 training complete.")

# Train the simple statistical detectors
ord_tester = ORDTester()
ord_tester.train(X_train_flat[y_train_flat == 0]) # Train on noise-only samples
trained_models['ORD'] = ord_tester


# In[6]:
# ==============================================================================
# FULL EVALUATION
# ==============================================================================
print("\n--- Starting Full Evaluation on Separate Test Set ---")
results = {model: {'pd_counts': np.zeros((NUM_FREQS, len(SIGNAL_SNRS))), 'fp_counts': np.zeros(NUM_FREQS)} for model in MODELS_TO_RUN}
total_eval_trials_per_snr = {}

with h5py.File(hdf5_filepath, 'r') as f:
    # --- Evaluate FP Rate ---
    noise_data_trials = f[f'snr_{NOISE_SNR}']['metrics'][:]
    total_eval_trials_per_snr[NOISE_SNR] = noise_data_trials.shape[0]
    
    for freq_idx in range(NUM_FREQS):
        std_features = noise_data_trials[:, freq_idx, :]
        ref_idx = (freq_idx + NUM_FREQS // 2) % NUM_FREQS
        aug_features = np.hstack([std_features, noise_data_trials[:, ref_idx, :]])
        
        if 'LGBM' in MODELS_TO_RUN: results['LGBM']['fp_counts'][freq_idx] = trained_models['LGBM'].predict(std_features).sum()
        if 'PPO' in MODELS_TO_RUN: actions, _ = trained_models['PPO'].predict(std_features, deterministic=True); actions[np.isnan(actions)]=0; results['PPO']['fp_counts'][freq_idx] = (actions > 0).sum()
        if 'PPO-Augmented' in MODELS_TO_RUN: actions, _ = trained_models['PPO-Augmented'].predict(aug_features, deterministic=True); actions[np.isnan(actions)]=0; results['PPO-Augmented']['fp_counts'][freq_idx] = (actions > 0).sum()
        if 'TD3' in MODELS_TO_RUN: actions, _ = trained_models['TD3'].predict(std_features, deterministic=True); actions[np.isnan(actions)]=0; results['TD3']['fp_counts'][freq_idx] = (actions > 0).sum()
        
        for w in ROLLING_WINDOW_SIZES:
            model_name = f'PPO-Rolling-W{w}'
            if model_name in MODELS_TO_RUN:
                rolling_features = create_rolling_features(noise_data_trials, w)[:, freq_idx, :]
                actions, _ = trained_models[model_name].predict(rolling_features, deterministic=True)
                actions[np.isnan(actions)]=0
                results[model_name]['fp_counts'][freq_idx] = (actions > 0).sum()

        ord_preds = trained_models['ORD'].predict(std_features)
        for det in ['MSC', 'CSM', 'TFG', 'TFL']:
            if det in MODELS_TO_RUN: results[det]['fp_counts'][freq_idx] = ord_preds[det].sum()

    # --- Evaluate PD Rate ---
    for snr_idx, snr in enumerate(tqdm(SIGNAL_SNRS, desc="Evaluating PD across SNRs")):
        signal_data_trials = f[f'snr_{snr}']['metrics'][:]
        total_eval_trials_per_snr[snr] = signal_data_trials.shape[0]
        
        for freq_idx in range(NUM_FREQS):
            std_features = signal_data_trials[:, freq_idx, :]
            ref_idx = (freq_idx + NUM_FREQS // 2) % NUM_FREQS
            aug_features = np.hstack([std_features, signal_data_trials[:, ref_idx, :]])

            if 'LGBM' in MODELS_TO_RUN: results['LGBM']['pd_counts'][freq_idx, snr_idx] = trained_models['LGBM'].predict(std_features).sum()
            if 'PPO' in MODELS_TO_RUN: actions, _ = trained_models['PPO'].predict(std_features, deterministic=True); actions[np.isnan(actions)]=0; results['PPO']['pd_counts'][freq_idx, snr_idx] = (actions > 0).sum()
            if 'PPO-Augmented' in MODELS_TO_RUN: actions, _ = trained_models['PPO-Augmented'].predict(aug_features, deterministic=True); actions[np.isnan(actions)]=0; results['PPO-Augmented']['pd_counts'][freq_idx, snr_idx] = (actions > 0).sum()
            if 'TD3' in MODELS_TO_RUN: actions, _ = trained_models['TD3'].predict(std_features, deterministic=True); actions[np.isnan(actions)]=0; results['TD3']['pd_counts'][freq_idx, snr_idx] = (actions > 0).sum()

            for w in ROLLING_WINDOW_SIZES:
                model_name = f'PPO-Rolling-W{w}'
                if model_name in MODELS_TO_RUN:
                    rolling_features = create_rolling_features(signal_data_trials, w)[:, freq_idx, :]
                    actions, _ = trained_models[model_name].predict(rolling_features, deterministic=True)
                    actions[np.isnan(actions)]=0
                    results[model_name]['pd_counts'][freq_idx, snr_idx] = (actions > 0).sum()

            ord_preds = trained_models['ORD'].predict(std_features)
            for det in ['MSC', 'CSM', 'TFG', 'TFL']:
                if det in MODELS_TO_RUN: results[det]['pd_counts'][freq_idx, snr_idx] = ord_preds[det].sum()

print("Evaluation complete.")


# In[7]:
# ==============================================================================
# PLOT 1: PER-FREQUENCY PERFORMANCE
# ==============================================================================
def wilson_score_interval(k, n, confidence_level=0.95):
    if n == 0 or k < 0: return 0.5, 0.0, 1.0
    z = norm.ppf(1 - (1 - confidence_level) / 2)
    phat = k / n
    term1 = phat + (z**2) / (2 * n)
    term2 = z * np.sqrt((phat * (1 - phat) / n) + (z**2) / (4 * n**2))
    denominator = 1 + (z**2) / n
    # FIX: Return rate, lower bound, upper bound
    return (phat, (term1 - term2) / denominator, (term1 + term2) / denominator)

print("\n--- Generating Per-Frequency Performance Plots ---")
for model_name in ['LGBM', 'PPO', 'PPO-Augmented', 'TD3'] + [f'PPO-Rolling-W{w}' for w in ROLLING_WINDOW_SIZES]:
    if model_name not in MODELS_TO_RUN: continue
    
    plt.figure(figsize=(12, 8))
    colors = plt.get_cmap('viridis', NUM_FREQS)
    
    for freq_idx in range(NUM_FREQS):
        pd_counts = results[model_name]['pd_counts'][freq_idx, :]
        pd_rates, lower_ci, upper_ci = [], [], []
        for snr_idx, snr in enumerate(SIGNAL_SNRS):
            n_trials = total_eval_trials_per_snr[snr]
            rate, low, high = wilson_score_interval(pd_counts[snr_idx], n_trials)
            pd_rates.append(rate)
            lower_ci.append(low)
            upper_ci.append(high)
            
        plt.plot(SIGNAL_SNRS, pd_rates, 'o-', color=colors(freq_idx), label=f"Freq {p_conf['signal_freqs'][freq_idx]} Hz")
        plt.fill_between(SIGNAL_SNRS, lower_ci, upper_ci, color=colors(freq_idx), alpha=0.1)

    plt.title(f'Per-Frequency PD vs. SNR for {model_name}', fontsize=16)
    plt.xlabel('SNR (dB)', fontsize=12)
    plt.ylabel('Probability of Detection (PD)', fontsize=12)
    plt.grid(True, linestyle=':')
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.ylim([-0.05, 1.05])
    plt.tight_layout()
    plt.show()


# In[8]:
# ==============================================================================
# PLOT 2: MEAN PERFORMANCE COMPARISON
# ==============================================================================
print("\n--- Generating Mean Performance Comparison Plot ---")
plt.figure(figsize=(14, 9))
colors = plt.get_cmap('tab10', len(MODELS_TO_RUN))

for i, model_name in enumerate(MODELS_TO_RUN):
    # Calculate Mean PD
    total_pd_counts = np.sum(results[model_name]['pd_counts'], axis=0)
    pd_rates, lower_ci, upper_ci = [], [], []
    for snr_idx, snr in enumerate(SIGNAL_SNRS):
        n_trials = total_eval_trials_per_snr[snr] * NUM_FREQS
        rate, low, high = wilson_score_interval(total_pd_counts[snr_idx], n_trials)
        pd_rates.append(rate)
        lower_ci.append(low)
        upper_ci.append(high)
        
    plt.plot(SIGNAL_SNRS, pd_rates, 'o-', color=colors(i), label=f'{model_name} PD')
    plt.fill_between(SIGNAL_SNRS, lower_ci, upper_ci, color=colors(i), alpha=0.1)
    
    # Calculate Mean FPR
    total_fp_counts = np.sum(results[model_name]['fp_counts'])
    n_fp_trials = total_eval_trials_per_snr[NOISE_SNR] * NUM_FREQS
    fp_rate, fp_low, fp_high = wilson_score_interval(total_fp_counts, n_fp_trials)
    plt.axhline(y=fp_rate, color=colors(i), linestyle='--', label=f'{model_name} FPR = {fp_rate:.3f}')

plt.title('Mean Detector Performance Comparison vs. SNR', fontsize=16)
plt.xlabel('SNR (dB)', fontsize=12)
plt.ylabel('Mean Probability', fontsize=12)
plt.grid(True, which='both', linestyle=':')
plt.ylim([-0.05, 1.05])
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()
plt.show()
```

```python

# In[8]:
# ==============================================================================
# PLOT 2: MEAN PERFORMANCE COMPARISON
# ==============================================================================
print("\n--- Generating Mean Performance Comparison Plot ---")
plt.figure(figsize=(20, 9))

# Define color and marker groups for model types
model_styles = {
    'RL': {'color': 'C0', 'markers': ['o', 's', 'v', '^', '<', '>']},
    'Supervised': {'color': 'C2', 'markers': ['D']},
    'Statistical': {'color': 'C3', 'markers': ['X', 'P', '*', 'H']}
}

rl_models = [m for m in MODELS_TO_RUN if 'PPO' in m or 'TD3' in m]
sup_models = [m for m in MODELS_TO_RUN if 'LGBM' in m]
stat_models = [m for m in MODELS_TO_RUN if m in ['MSC', 'CSM', 'TFG', 'TFL']]

def get_style(model_name, index):
    if model_name in rl_models: return model_styles['RL']['color'], model_styles['RL']['markers'][index % len(model_styles['RL']['markers'])]
    if model_name in sup_models: return model_styles['Supervised']['color'], model_styles['Supervised']['markers'][index % len(model_styles['Supervised']['markers'])]
    if model_name in stat_models: return model_styles['Statistical']['color'], model_styles['Statistical']['markers'][index % len(model_styles['Statistical']['markers'])]
    return 'k', '.' # Default

for i, model_name in enumerate(MODELS_TO_RUN):
    color, marker = get_style(model_name, i)
    
    # Calculate Mean PD
    total_pd_counts = np.sum(results[model_name]['pd_counts'], axis=0)
    pd_rates, lower_ci, upper_ci = [], [], []
    for snr_idx, snr in enumerate(SIGNAL_SNRS):
        n_trials = total_eval_trials_per_snr[snr] * NUM_FREQS
        rate, low, high = wilson_score_interval(total_pd_counts[snr_idx], n_trials,confidence_level=0.99)
        pd_rates.append(rate)
        lower_ci.append(low)
        upper_ci.append(high)
        
    plt.plot(SIGNAL_SNRS, pd_rates, marker=marker, linestyle='-', color=color, label=f'{model_name} PD',linewidth= 0.9)
    plt.fill_between(SIGNAL_SNRS, lower_ci, upper_ci, color=color, alpha=0.4)
    
    # Calculate Mean FPR
    total_fp_counts = np.sum(results[model_name]['fp_counts'])
    n_fp_trials = total_eval_trials_per_snr[NOISE_SNR] * NUM_FREQS
    fp_rate, _, _ = wilson_score_interval(total_fp_counts, n_fp_trials)
    plt.axhline(y=fp_rate, color=color, linestyle='--', label=f'{model_name} FPR = {fp_rate:.3f}')

plt.title('Mean Detector Performance Comparison vs. SNR', fontsize=16)
plt.xlabel('SNR (dB)', fontsize=12)
plt.ylabel('Mean Probability', fontsize=12)
plt.grid(True, which='both', linestyle=':')
plt.ylim([-0.05, 1.05])
plt.legend(loc='upper left'
        #    , bbox_to_anchor=(1, 0.5)
           )
plt.tight_layout()
plt.show()
```
