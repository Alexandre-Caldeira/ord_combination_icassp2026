```python
# coding: utf-8

# # Treinamento Iterativo e Avaliação de Performance em Larga Escala de Detectores
# 
# **Autor:** Engenheiro de Machine Learning Sênior
# **Data:** 07 de agosto de 2025
# 
# ## Objetivo
# Este notebook implementa um fluxo de treinamento e avaliação aninhado e iterativo.
# 1.  **Curva de Aprendizado (Treino):** Analisa como a performance do modelo evolui com o aumento do volume de dados de treinamento.
# 2.  **Convergência de Métricas (Avaliação):** Analisa como as métricas de performance (média, variância) se estabilizam com o aumento do número de amostras de avaliação.

# ## 1. Configuração e Bibliotecas
# 
# Importação das bibliotecas e definição dos parâmetros do experimento.

# In[1]:


import numpy as np
import h5py
import yaml
import os
import pickle
import lightgbm as lgb
import gymnasium as gym
from gymnasium import spaces
from stable_baselines3 import PPO
from tqdm.notebook import tqdm
import pandas as pd
import matplotlib.pyplot as plt
import warnings

# Ignorar avisos de nomes de features do LightGBM
warnings.filterwarnings("ignore", category=UserWarning, message="X does not have valid feature names")

# --- Parâmetros do Experimento (AJUSTÁVEL) ---
# Pontos de parada para o NÚMERO DE AMOSTRAS DE TREINO
TRAINING_SAMPLE_STOPS = [1000, 10000]

# Pontos de parada para o NÚMERO DE TRIALS DE AVALIAÇÃO POR SNR
EVAL_TRIAL_STOPS = [1000]

# --- Parâmetros de Treinamento ---
MODELS_TO_TRAIN_AND_EVAL = ['PPO-Rolling-Avg-W18', 'LGBM', 'MSC', 'TFG']
ROLLING_WINDOW_SIZE = 18
RL_TRAINING_EPOCHS = 10  # Número de passagens completas sobre os dados de treino para o PPO
LGBM_N_ESTIMATORS = 100
NEGATIVE_REWARD = -2.5

# --- Caminhos e Configurações ---
config_path = '../config.yml'
with open(config_path, 'r') as f:
    config = yaml.safe_load(f)

HDF5_TRAIN_FILEPATH = '../data/train.hdf5'
HDF5_EVAL_FILEPATH = '../data/eval.hdf5'

p_conf = config['parameters']
NUM_SIGNAL_FREQS = len(p_conf['signal_freqs'])

# ADICIONADO: Defina a lista de SNRs para usar EXCLUSIVAMENTE no treino
SNRS_FOR_TRAINING = [5.0, 0.0, -5.0] # <-- NOVO: Lista de SNRs para o treino

# Mantém todos os SNRs para a fase de AVALIAÇÃO, para testar a generalização
ALL_SNRS = p_conf['snr_range']

# Carregar um número suficiente de dados de treino e avaliação para o maior "stop"
TRAIN_TRIALS_PER_SNR =  max(TRAINING_SAMPLE_STOPS)
MAX_EVAL_TRIALS_PER_SNR = max(EVAL_TRIAL_STOPS)

print(f"Pontos de parada para treinamento (nº de amostras): {TRAINING_SAMPLE_STOPS}")
print(f"Pontos de parada para avaliação (nº de trials por SNR): {EVAL_TRIAL_STOPS}")
print(f"Modelos a serem avaliados: {MODELS_TO_TRAIN_AND_EVAL}")


# ## 2. Funções de Carga e Preparação de Dados
# 
# Funções para carregar os dados de treino e avaliação.

# In[2]:


def create_rolling_avg_features(trial_data, window_size):
    """Cria características de média móvel para um conjunto de dados de trials."""
    n_trials, n_steps, n_features = trial_data.shape
    rolling_features = np.zeros_like(trial_data)
    for i in range(n_steps):
        start_idx = max(0, i - window_size + 1)
        rolling_features[:, i, :] = np.mean(trial_data[:, start_idx:i+1, :], axis=1)
    return rolling_features

def create_cumulative_avg_features(trial_data):
    """Cria características de média cumulativa (expanding mean) de forma eficiente."""
    cumsum_features = np.cumsum(trial_data, axis=1)
    denominators = np.arange(1, trial_data.shape[1] + 1).reshape(1, -1, 1)
    cumulative_features = cumsum_features / denominators
    return cumulative_features

def load_and_prepare_full_training_data(filepath, snr_levels, trials_per_snr, num_signal_freqs, window_size):
    """Carrega o conjunto de dados de treino COMPLETO e prepara todas as variantes."""
    X_sig_trials, X_noise_trials = [], []
    with h5py.File(filepath, 'r') as f:
        for snr in snr_levels: # Note: This now iterates over SNRS_FOR_TRAINING
            metrics_data = f[f'snr_{snr}']['metrics'][:trials_per_snr, :, :]
            X_sig_trials.append(metrics_data[:, :num_signal_freqs, :])
            X_noise_trials.append(metrics_data[:, num_signal_freqs:, :])

    X_sig = np.vstack(X_sig_trials)
    X_noise = np.vstack(X_noise_trials)
    n_features = X_sig.shape[-1]
    
    X_flat_sig = X_sig.reshape(-1, n_features)
    X_flat_noise = X_noise.reshape(-1, n_features)
    X_train_flat = np.vstack([X_flat_sig, X_flat_noise])
    y_train_flat = np.concatenate([np.ones(len(X_flat_sig)), np.zeros(len(X_flat_noise))])
    p_flat = np.random.permutation(len(X_train_flat))
    X_train_flat, y_train_flat = X_train_flat[p_flat], y_train_flat[p_flat]

    X_ordered = np.vstack([X_sig, X_noise])
    y_ordered_labels = np.vstack([np.ones((len(X_sig), num_signal_freqs)), np.zeros((len(X_noise), num_signal_freqs))])
    p_ord = np.random.permutation(len(X_ordered))
    X_ordered, y_ordered_labels = X_ordered[p_ord], y_ordered_labels[p_ord]
    
    X_rolling_trials = create_rolling_avg_features(X_ordered, window_size)
    X_cumulative_trials = create_cumulative_avg_features(X_ordered)
    
    X_train_rolling = X_rolling_trials.reshape(-1, n_features)
    X_train_cumulative = X_cumulative_trials.reshape(-1, n_features)
    y_train_rolling = y_ordered_labels.flatten()
    
    X_train_flat_combined = np.concatenate([X_train_flat, X_train_cumulative], axis=1)
    X_train_rolling_combined = np.concatenate([X_train_rolling, X_train_cumulative], axis=1)

    datasets = {
        'flat_combined': (X_train_flat_combined, y_train_flat),
        'rolling_cumulative_combined': (X_train_rolling_combined, y_train_rolling)
    }
    return datasets

def load_evaluation_data_for_snr(filepath, snr, trials_per_snr, num_signal_freqs):
    """Carrega e separa os dados de sinal e ruído para um único valor de SNR do arquivo de EVAL."""
    with h5py.File(filepath, 'r') as f:
        metrics_data = f[f'snr_{snr}']['metrics'][:trials_per_snr, :, :]
        signal_trials = metrics_data[:, :num_signal_freqs, :]
        noise_trials = metrics_data[:, num_signal_freqs:, :]
    return signal_trials, noise_trials

print("--- Carregando Conjunto de Dados de Treino Completo de 'train.hdf5' ---")
full_train_datasets = load_and_prepare_full_training_data(
    HDF5_TRAIN_FILEPATH, SNRS_FOR_TRAINING, TRAIN_TRIALS_PER_SNR, NUM_SIGNAL_FREQS, ROLLING_WINDOW_SIZE
)
print(f"Dados de treino carregados APENAS para os SNRs: {SNRS_FOR_TRAINING}")


# ## 3. Ambiente Gymnasium e Funções Auxiliares

# In[3]:


class SignalDetectionEnv(gym.Env):
    """Ambiente Gymnasium para os agentes de RL."""
    def __init__(self, features, labels, negative_reward=-1.0):
        super().__init__()
        self.features, self.labels = features, labels
        self.negative_reward = negative_reward
        self.current_step = 0
        self.action_space = spaces.Box(low=-1, high=1, shape=(1,), dtype=np.float32)
        self.observation_space = spaces.Box(low=-np.inf, high=np.inf, shape=(features.shape[1],), dtype=np.float32)

    def step(self, action):
        predicted_label = 1 if action[0] > 0 else 0
        reward = 1.0 if predicted_label == self.labels[self.current_step] else self.negative_reward
        self.current_step += 1
        terminated = self.current_step >= len(self.features)
        next_obs = self.features[self.current_step] if not terminated else np.zeros(self.observation_space.shape)
        return next_obs, reward, terminated, False, {}

    def reset(self, seed=None, options=None):
        super().reset(seed=seed)
        self.current_step = 0
        return self.features[0], {}

class ORDTester:
    """Classe para treinamento e predição dos detectores estatísticos."""
    def __init__(self, desired_alpha=0.05):
        self.desired_alpha = desired_alpha
        self.feature_names = ['MSC', 'CSM', 'TFG', 'TFL', 'SNR_meas', 'MAG_freq', 'PHI_freq']
        self.thresholds = {}
    def train(self, noise_features):
        for i, key in enumerate(self.feature_names):
            if noise_features.shape[0] > 0:
                self.thresholds[key] = np.percentile(noise_features[:, i], (1 - self.desired_alpha) * 100)
    def predict(self, features):
        decisions = {}
        for i, name in enumerate(self.feature_names):
            if name in self.thresholds:
                decisions[name] = (features[:, i] > self.thresholds[name]).astype(int)
            else:
                decisions[name] = np.zeros(features.shape[0], dtype=int)
        return decisions

def predict_lgbm(model, features_flat): return model.predict(features_flat)

def predict_ppo(model, features_flat):
    preds, _ = model.predict(features_flat, deterministic=True)
    preds[np.isnan(preds)] = 0
    return (preds > 0).astype(int)

def predict_statistical(model, features_flat, detector_name):
    decisions = model.predict(features_flat)
    return decisions[detector_name]

def calculate_statistics(predictions_per_channel):
    per_channel_metric = np.mean(predictions_per_channel, axis=0)
    stats = {
        'mean': np.mean(per_channel_metric), 'variance': np.var(per_channel_metric),
        'percentiles': {
            'p1': np.percentile(per_channel_metric, 1), 'p99': np.percentile(per_channel_metric, 99)
        }
    }
    return stats


# ## 4. Loop Principal de Treinamento e Avaliação Iterativa

# In[4]:


all_results_over_time = []
policy_kwargs = dict(net_arch=dict(pi=[64, 64], vf=[64, 64]))

for training_size in tqdm(TRAINING_SAMPLE_STOPS, desc="Progresso Total do Experimento"):
    print(f"\n{'='*80}")
    print(f"INICIANDO ETAPA DE TREINO: {training_size} amostras.")
    print(f"{'='*80}")
    
    X_train_flat_slice = full_train_datasets['flat_combined'][0][:training_size]
    y_train_flat_slice = full_train_datasets['flat_combined'][1][:training_size]
    X_train_rolling_slice = full_train_datasets['rolling_cumulative_combined'][0][:training_size]
    y_train_rolling_slice = full_train_datasets['rolling_cumulative_combined'][1][:training_size]

    trained_models_step = {}
    
    if 'LGBM' in MODELS_TO_TRAIN_AND_EVAL:
        print("Treinando LGBM...")
        model_lgbm = lgb.LGBMClassifier(n_estimators=LGBM_N_ESTIMATORS, random_state=42)
        model_lgbm.fit(X_train_flat_slice, y_train_flat_slice)
        trained_models_step['LGBM'] = model_lgbm

    if 'PPO-Rolling-Avg-W18' in MODELS_TO_TRAIN_AND_EVAL:
        # --- Stage 1: Standard Training (Discovery) ---
        print("Treinando PPO-Rolling-Avg-W18 (Stage 1: Discovery)...")
        env = SignalDetectionEnv(X_train_rolling_slice, y_train_rolling_slice, NEGATIVE_REWARD)
        
        initial_learning_rate = 3e-4
        initial_ent_coef = 0.0
        
        model_ppo = PPO("MlpPolicy", 
                      env, 
                      policy_kwargs=policy_kwargs, 
                      n_steps=min(1024, training_size), 
                      learning_rate=initial_learning_rate,
                      ent_coef=initial_ent_coef,
                      verbose=0)
        
        total_timesteps = training_size * RL_TRAINING_EPOCHS
        model_ppo.learn(total_timesteps=total_timesteps, progress_bar=True)

        # --- Stage 2: Fine-Tuning (Hardening) ---
        print("Iniciando PPO (Stage 2: Hardening/Fine-tuning)...")
        
        FINETUNE_LEARNING_RATE = 3e-5
        FINETUNE_ENT_COEF = 0.0
        FINETUNE_EPOCHS = 4
        
        model_ppo.learning_rate = FINETUNE_LEARNING_RATE
        model_ppo.ent_coef = FINETUNE_ENT_COEF
        
        finetune_timesteps = training_size * FINETUNE_EPOCHS
        model_ppo.learn(total_timesteps=finetune_timesteps, 
                        progress_bar=True, 
                        reset_num_timesteps=False)
        
        trained_models_step['PPO-Rolling-Avg-W18'] = model_ppo

    if 'MSC' in MODELS_TO_TRAIN_AND_EVAL or 'TFG' in MODELS_TO_TRAIN_AND_EVAL:
        print("Treinando Detectores Estatísticos (ORD)...")
        ord_tester = ORDTester()
        ord_tester.train(X_train_flat_slice[y_train_flat_slice == 0])
        trained_models_step['MSC'] = ord_tester
        trained_models_step['TFG'] = ord_tester
        
    print(f"\n--- Iniciando Avaliação para modelos treinados com {training_size} amostras ---")
    
    for eval_trials in tqdm(EVAL_TRIAL_STOPS, desc="Etapas de Avaliação", leave=False):
        step_results = {}
        for model_name in MODELS_TO_TRAIN_AND_EVAL:
            if model_name not in trained_models_step: continue
            
            step_results[model_name] = {'TD': {}, 'FPR': {}}
            for snr in ALL_SNRS: # Evaluation still runs on ALL SNRs
                signal_trials, noise_trials = load_evaluation_data_for_snr(HDF5_EVAL_FILEPATH, snr, eval_trials, NUM_SIGNAL_FREQS)
                
                if signal_trials.shape[0] == 0 or noise_trials.shape[0] == 0: continue
                
                cumulative_pd_trials = create_cumulative_avg_features(signal_trials)
                cumulative_fpr_trials = create_cumulative_avg_features(noise_trials)

                n_features = signal_trials.shape[-1]
                
                if model_name == 'PPO-Rolling-Avg-W18':
                    rolling_pd = create_rolling_avg_features(signal_trials, ROLLING_WINDOW_SIZE).reshape(-1, n_features)
                    cumulative_pd = cumulative_pd_trials.reshape(-1, n_features)
                    features_pd = np.concatenate([rolling_pd, cumulative_pd], axis=1)
                    preds_pd = predict_ppo(trained_models_step[model_name], features_pd)
                    
                    rolling_fpr = create_rolling_avg_features(noise_trials, ROLLING_WINDOW_SIZE).reshape(-1, n_features)
                    cumulative_fpr = cumulative_fpr_trials.reshape(-1, n_features)
                    features_fpr = np.concatenate([rolling_fpr, cumulative_fpr], axis=1)
                    preds_fpr = predict_ppo(trained_models_step[model_name], features_fpr)
                else:
                    original_pd = signal_trials.reshape(-1, n_features)
                    cumulative_pd = cumulative_pd_trials.reshape(-1, n_features)
                    features_pd = np.concatenate([original_pd, cumulative_pd], axis=1)

                    original_fpr = noise_trials.reshape(-1, n_features)
                    cumulative_fpr = cumulative_fpr_trials.reshape(-1, n_features)
                    features_fpr = np.concatenate([original_fpr, cumulative_fpr], axis=1)

                    if model_name == 'LGBM':
                        preds_pd = predict_lgbm(trained_models_step[model_name], features_pd)
                        preds_fpr = predict_lgbm(trained_models_step[model_name], features_fpr)
                    else:
                        preds_pd = predict_statistical(trained_models_step[model_name], features_pd, model_name)
                        preds_fpr = predict_statistical(trained_models_step[model_name], features_fpr, model_name)

                step_results[model_name]['TD'][snr] = calculate_statistics(preds_pd.reshape(eval_trials, -1))
                step_results[model_name]['FPR'][snr] = calculate_statistics(preds_fpr.reshape(eval_trials, -1))

        for model, metrics in step_results.items():
            for snr, stats in metrics['TD'].items():
                all_results_over_time.append({'Amostras Treino': training_size, 'Trials Avaliação': eval_trials, 'Modelo': model, 'Métrica': 'TD', 'SNR': snr, 'Média': stats['mean'], 'Variância': stats['variance'], 'P1': stats['percentiles']['p1'], 'P99': stats['percentiles']['p99']})
            for snr, stats in metrics['FPR'].items():
                all_results_over_time.append({'Amostras Treino': training_size, 'Trials Avaliação': eval_trials, 'Modelo': model, 'Métrica': 'FPR', 'SNR': snr, 'Média': stats['mean'], 'Variância': stats['variance'], 'P1': stats['percentiles']['p1'], 'P99': stats['percentiles']['p99']})

    df_final_eval_step = pd.DataFrame([r for r in all_results_over_time if r['Amostras Treino'] == training_size and r['Trials Avaliação'] == MAX_EVAL_TRIALS_PER_SNR])
    if not df_final_eval_step.empty:
        print(f"\n--- RESULTADOS FINAIS PARA {training_size} AMOSTRAS DE TREINO (Avaliados com {MAX_EVAL_TRIALS_PER_SNR} trials) ---")
        print("--- Média da Taxa de Detecção (TD) vs. SNR ---")
        display(df_final_eval_step[df_final_eval_step['Métrica'] == 'TD'].pivot(index='SNR', columns='Modelo', values='Média'))
        print("\n--- Média da Taxa de Falsos Positivos (FPR) vs. SNR ---")
        display(df_final_eval_step[df_final_eval_step['Métrica'] == 'FPR'].pivot(index='SNR', columns='Modelo', values='Média'))

print("\n\n--- EXPERIMENTO COMPLETO CONCLUÍDO ---")


# ## 5. Análise Final e Visualização

# In[5]:


final_df = pd.DataFrame(all_results_over_time)

print("\n\n--- Evolução da Média de TD (em SNR= -10dB) vs. Amostras de Treino e Trials de Avaliação ---")
display(final_df[(final_df['Métrica'] == 'TD') & (final_df['SNR'] == -10.0)].pivot_table(index='Amostras Treino', columns='Trials Avaliação', values='Média', aggfunc='mean'))

print("\n\n--- Evolução da Variância de TD (em SNR= -10dB) vs. Amostras de Treino e Trials de Avaliação ---")
display(final_df[(final_df['Métrica'] == 'TD') & (final_df['SNR'] == -10.0)].pivot_table(index='Amostras Treino', columns='Trials Avaliação', values='Variância', aggfunc='mean'))

print("\n\n--- Evolução da Média de FPR (em SNR= -10dB) vs. Amostras de Treino e Trials de Avaliação ---")
display(final_df[(final_df['Métrica'] == 'FPR') & (final_df['SNR'] == -10.0)].pivot_table(index='Amostras Treino', columns='Trials Avaliação', values='Média', aggfunc='mean'))

# ## 6. Gráficos de Performance com Intervalo de Confiança

# In[6]:


def plot_performance_curves_with_ci(df, training_stop, eval_stop):
    """
    Plota as curvas de TD e FPR vs. SNR com intervalos de confiança de 98% (P1 a P99).
    """
    df_plot = df[(df['Amostras Treino'] == training_stop) & (df['Trials Avaliação'] == eval_stop)]
    
    if df_plot.empty:
        print(f"Nenhum dado de avaliação encontrado para {training_stop} amostras de treino e {eval_stop} trials de avaliação.")
        return

    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 16), sharex=True)
    fig.suptitle(f'Performance dos Modelos (Treino: {training_stop} amostras, Avaliação: {eval_stop} trials)', fontsize=18)

    colors = plt.cm.viridis(np.linspace(0, 1, len(MODELS_TO_TRAIN_AND_EVAL)))
    model_colors = {model: color for model, color in zip(MODELS_TO_TRAIN_AND_EVAL, colors)}

    for model in MODELS_TO_TRAIN_AND_EVAL:
        model_data = df_plot[(df_plot['Modelo'] == model) & (df_plot['Métrica'] == 'TD')].sort_values('SNR')
        if not model_data.empty:
            ax1.plot(model_data['SNR'], model_data['Média'], label=f'{model} Média TD', color=model_colors[model], marker='o')
            ax1.fill_between(model_data['SNR'], model_data['P1'], model_data['P99'], color=model_colors[model], alpha=0.2, label=f'{model} 98% CI')

    ax1.set_ylabel('Taxa de Detecção (TD)', fontsize=14)
    ax1.set_title('Taxa de Detecção vs. SNR', fontsize=16)
    ax1.grid(True, linestyle=':')
    ax1.legend()
    ax1.set_ylim(-0.05, 1.05)

    for model in MODELS_TO_TRAIN_AND_EVAL:
        model_data = df_plot[(df_plot['Modelo'] == model) & (df_plot['Métrica'] == 'FPR')].sort_values('SNR')
        if not model_data.empty:
            ax2.plot(model_data['SNR'], model_data['Média'], label=f'{model} Média FPR', color=model_colors[model], marker='x')
            ax2.fill_between(model_data['SNR'], model_data['P1'], model_data['P99'], color=model_colors[model], alpha=0.2, label=f'{model} 98% CI')

    ax2.set_xlabel('SNR (dB)', fontsize=14)
    ax2.set_ylabel('Taxa de Falsos Positivos (FPR)', fontsize=14)
    ax2.set_title('Taxa de Falsos Positivos vs. SNR', fontsize=16)
    ax2.grid(True, linestyle=':')
    ax2.legend()
    ax2.set_ylim(bottom=-0.01)
    
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

# Plotar os resultados para as etapas de treino desejadas
STAGES_TO_PLOT = [
    (1000, 1000),
    (10000, 1000),
    (20000, 1000),
    (50000, 1000)
]
print("\n\n--- Gráficos de Performance para Etapas Selecionadas ---")
for train_stop, eval_stop in STAGES_TO_PLOT:
    plot_performance_curves_with_ci(final_df, train_stop, eval_stop)
```

```python

# ## 6. Fase 2: Comparação das Abordagens de Ajuste Fino

# In[6]:
EXPERIMENTAL_FEATURES_FILE = '../data/experimental_features.hdf5'
MODELS_DIR = './models'
CHECKPOINT_TRAINING_SIZE = max(TRAINING_SAMPLE_STOPS)
NEGATIVE_REWARD_EXP = -5.0

print(f"\n\n{'='*80}\n--- INICIANDO FASE 2: AJUSTE FINO E AVALIAÇÃO EM DADOS EXPERIMENTAIS ---\n{'='*80}")

if not os.path.exists(EXPERIMENTAL_FEATURES_FILE):
    print(f"AVISO: Arquivo de dados experimentais '{EXPERIMENTAL_FEATURES_FILE}' não encontrado. Pulando a Fase 2.")
else:
    with h5py.File(EXPERIMENTAL_FEATURES_FILE, 'r') as f: volunteers = list(f.keys())
    results = {model: {k:[] for k in ['baseline_tpr','baseline_fpr','finetune_tpr','finetune_fpr','retrain_tpr','retrain_fpr']} for model in MODELS_TO_TRAIN_AND_EVAL}

    def evaluate_on_experimental(model, X_raw, model_name):
        n_features = X_raw.shape[-1]
        X_orig_flat = X_raw.reshape(-1, n_features)
        X_rolling_flat = create_rolling_avg_features(X_raw, ROLLING_WINDOW_SIZE).reshape(-1, n_features)
        X_cumulative_flat = create_cumulative_avg_features(X_raw).reshape(-1, n_features)

        if model_name == 'PPO-Rolling-Avg-W18':
            features = np.concatenate([X_rolling_flat, X_cumulative_flat], axis=1); preds = predict_ppo(model, features)
        elif model_name == 'LGBM':
            features = np.concatenate([X_orig_flat, X_cumulative_flat], axis=1); preds = predict_lgbm(model, features)
        else:
            features = np.concatenate([X_orig_flat, X_cumulative_flat], axis=1); preds = predict_statistical(model, features, model_name)
        
        preds_reshaped = preds.reshape(X_raw.shape[0], -1)
        return np.mean(preds_reshaped[:, :NUM_SIGNAL_FREQS]), np.mean(preds_reshaped[:, NUM_SIGNAL_FREQS:])

    # ADIÇÃO: Envolve o loop principal da Fase 2 com tqdm para ETA.
    for tune_volunteer in tqdm(volunteers, desc="Progresso Folds Fine-Tuning"):
        print(f"\n--- Fold: Usando '{tune_volunteer}' para ajuste ---")
        
        with h5py.File(EXPERIMENTAL_FEATURES_FILE, 'r') as f:
            X_tune_raw = np.vstack([f[tune_volunteer][intensity]['metrics'][:] for intensity in f[tune_volunteer].keys()])
            X_test_list = []
            test_volunteers = [v for v in volunteers if v != tune_volunteer]
            for v_test in test_volunteers:
                for intensity in f[v_test].keys():
                    X_test_list.append(f[v_test][intensity]['metrics'][:])
        
        if not X_test_list:
            print("  - AVISO: Nenhum dado de teste encontrado para este fold. Pulando.")
            continue
        X_test_raw = np.vstack(X_test_list)

        for model_name in MODELS_TO_TRAIN_AND_EVAL:
            model_path = os.path.join(MODELS_DIR, f'{model_name}_{CHECKPOINT_TRAINING_SIZE}_samples.pkl')
            if not os.path.exists(model_path): continue
            
            print(f"  Avaliando modelo: {model_name}")
            with open(model_path, 'rb') as f: base_model = pickle.load(f)

            tpr, fpr = evaluate_on_experimental(base_model, X_test_raw, model_name)
            results[model_name]['baseline_tpr'].append(tpr); results[model_name]['baseline_fpr'].append(fpr)
            # ADIÇÃO: Verbosidade para o resultado do baseline
            print(f"    - Baseline : TPR={100*tpr:.3f}%, FPR={100*fpr:.3f}%")


            n_features = X_tune_raw.shape[-1]
            X_tune_flat_orig = X_tune_raw.reshape(-1, n_features)
            y_tune_flat = np.tile(np.concatenate([np.ones(NUM_SIGNAL_FREQS), np.zeros(NUM_SIGNAL_FREQS)]), X_tune_raw.shape[0])
            X_tune_cumulative_flat = create_cumulative_avg_features(X_tune_raw).reshape(-1, n_features)
            
            ft_model, re_model = None, None
            if model_name == 'LGBM':
                X_tune_lgbm = np.concatenate([X_tune_flat_orig, X_tune_cumulative_flat], axis=1)
                re_model = lgb.LGBMClassifier(n_estimators=LGBM_N_ESTIMATORS, random_state=42).fit(X_tune_lgbm, y_tune_flat)
                ft_model = lgb.LGBMClassifier(n_estimators=LGBM_N_ESTIMATORS, random_state=42).fit(X_tune_lgbm, y_tune_flat, init_model=base_model)
            elif model_name == 'PPO-Rolling-Avg-W18':
                X_tune_rolling_flat = create_rolling_avg_features(X_tune_raw, ROLLING_WINDOW_SIZE).reshape(-1, n_features)
                X_tune_ppo = np.concatenate([X_tune_rolling_flat, X_tune_cumulative_flat], axis=1)
                env = SignalDetectionEnv(X_tune_ppo, y_tune_flat, NEGATIVE_REWARD_EXP)
                re_model = PPO("MlpPolicy", env, policy_kwargs=policy_kwargs, n_steps=min(1024, len(X_tune_ppo)), verbose=1).learn(total_timesteps=len(X_tune_ppo) * 1)
                ft_model = re_model
            else:
                X_tune_ord = np.concatenate([X_tune_flat_orig, X_tune_cumulative_flat], axis=1)
                re_model = ORDTester(); re_model.train(X_tune_ord[y_tune_flat == 0]); ft_model = re_model

            if ft_model:
                tpr, fpr = evaluate_on_experimental(ft_model, X_test_raw, model_name)
                results[model_name]['finetune_tpr'].append(tpr); results[model_name]['finetune_fpr'].append(fpr)
                # ADIÇÃO: Verbosidade para o resultado do fine-tune
                print(f"    - Fine-tune: TPR={100*tpr:.3f}%, FPR={100*fpr:.3f}%")

            if re_model:
                tpr, fpr = evaluate_on_experimental(re_model, X_test_raw, model_name)
                results[model_name]['retrain_tpr'].append(tpr); results[model_name]['retrain_fpr'].append(fpr)
                # ADIÇÃO: Verbosidade para o resultado do re-treino
                print(f"    - Retrain  : TPR={100*tpr:.3f}%, FPR={100*fpr:.3f}%")
    
    print("\n\n--- RESULTADOS FINAIS DO AJUSTE FINO (MÉDIA +/- DESVIO PADRÃO) ---")
    for model_name in MODELS_TO_TRAIN_AND_EVAL:
        print(f"\nResultados para: {model_name}")
        for approach in ['baseline', 'finetune', 'retrain']:
            tpr_mean, tpr_std = np.mean(results[model_name][f'{approach}_tpr']), np.std(results[model_name][f'{approach}_tpr'])
            fpr_mean, fpr_std = np.mean(results[model_name][f'{approach}_fpr']), np.std(results[model_name][f'{approach}_fpr'])
            if not np.isnan(tpr_mean):
                print(f"  {approach.capitalize():<13}: TPR = {tpr_mean:.3f} ± {tpr_std:.3f} | FPR = {fpr_mean:.3f} ± {fpr_std:.3f}")
```
